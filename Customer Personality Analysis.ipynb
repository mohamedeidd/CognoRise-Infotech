# clustring customers

```python
df.drop(axis = 1 ,columns=['Kidhome','Teenhome','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2' , 'ID' ,'Year_Birth' ,'Dt_Customer' ,'Complain', 'Z_CostContact','Z_Revenue'  ] , inplace=True)
```

```python
df.columns
```

```python
df
```

```python
Education = pd.get_dummies(df.Education ,drop_first=True )
Marital_Status = pd.get_dummies(df.Marital_Status , drop_first=True)
```

```python
Education
```

```python
Marital_Status
```

```python
df.drop(axis=1 , columns=['Education', 'Marital_Status'] , inplace=True)
```

```python
df = pd.concat([df, Education, Marital_Status] , axis =1 )
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard = StandardScaler()
scaled_data = standard.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42 , max_iter=2000)
```

```python
df['cluster']=kmeans.fit_predict(df)
```

```python
df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)


df['PCA1'] = pca_components[:, 0]
df['PCA2'] = pca_components[:, 1]


plt.figure(figsize=(10, 7))
sns.scatterplot(x='PCA1', y='PCA2', hue='cluster', palette='Set1', data=df, s=100, alpha=0.7)
plt.title('Customer Segments based on Spending Behavior')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.show()
```
